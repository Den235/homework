```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}


def sayHI():
    print("Привет, выбери нужную функцию:")
    print("Если ты хочешь узнать владельца документа по его номеру - введи p")
    print("Если ты хочешь по номеру документа узнать на какой полке он хранится - введи s")
    print("Если ты хочешь увидеть полную информацию по всем документам - введи l")
    print("Если ты хочешь добавить новую полку - введи ads")
    print("Если ты хочешь удалить существующую полку из данных (только если она пустая!) - введи ds")
    
    option = ""
    while option != "q":
        option = input("Введите параметр: ")
        if(option == "p"):
            return p()
        elif (option == "s"):
            return print("Документ хранится на полке: " + s())
        elif (option == "l"):
            return l()
        elif (option == "ads"):
            return ads()
        elif (option == "ds"):
            return ds()
      
    return ('Пока')


def p():
    number = input("Введите номер документа ")
    for i in documents:
        if i['number'] == number:
            return print("Владелец документа: " + i['name'])
    return print("Документ не найден в базе")
    
    
    
def s():
    number = input("Введите номер документа ")
    for i in directories:
        if number in directories[i]:
            return i
    return print("Документ не найден в базе")

def l():
    for i in documents:
        print("№: " + i['number'] + ", " + "тип: " + i['type'] + ", " + "владелец: " + i['name'] + ", " + "полка хранения: " + s(i['number']))

        
def ads():
    number = input("Введите номер полки: ")
    if number in directories:
        return print("Такая полка уже существует. Текущий перечень полок: " + ','.join(list(directories.keys())))
    else:
        directories[number] = []
        print("Полка добавлена. Текущий перечень полок: " + ','.join(list(directories.keys())))
    
def ds():
    number = input("Введите номер полки: ") 
    if(number not in directories.keys()):
        print("Такой полки не существует. Текущий перечень полок: " + ','.join(list(directories.keys()))) 
    elif(len(directories[number]) == 0):
        del(directories[number])
        print("Полка удалена. Текущий перечень полок: " + ','.join(list(directories.keys())))
    else:
        print("На полке есть документа, удалите их перед удалением полки.Текущий перечень полок: " + ','.join(list(directories.keys()))) 
    
    
    
sayHI()






```
